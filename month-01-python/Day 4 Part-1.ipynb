```python
# loop range()
x=range(0,5)
for i in x:
    print(i)
```

```python
# some more examples
fruits=['mango','apple','orange','banana']
for i in range(len(fruits)):
    print(f'{i} and {fruits[i]}')

# we can write range(start,stop,step)
number = '0123456789'
for i in range(0,len(number),2):  # this step will make i as to step by 2 everytime it iterates
    print(i,number[i])
```

```python
# lets learn about break statement
# if we use the break ,the below statements in the loop will not execute at all 
fruits=['mango','apple','orange','banana']
for i in fruits:
    if i== 'orange':
        break
    print(i)
```

```python
# The continue statement skips the current iteration of the loop and continues with the next iteration. For example,
fruits=['mango','apple','orange','banana']
for i in fruits:
    if i== 'orange': continue
    print(i)
```

```python
#nested loop
attributes = ['Electric', 'Fast']
cars = ['Tesla', 'Porsche', 'Mercedes']
for i in attributes:
    for j in cars:
        print(i,j)
    print('------------')
```

```python
# If we don't intend to use items of sequence inside the body of a loop, it is clearer to use the _ (underscore) as the loop variable. For example,

for _ in range(4):
    print(_,'hi')
```
