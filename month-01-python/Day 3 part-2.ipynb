```python
#lets learn about If else statement
number = int(input("Enter a number:"))
if number>0 :
     print( f'{number} is a positive number.')  # this is a good way to write print statement

#lets learn about indentation
#We usually use four spaces for indentation in Python, although any number of spaces works as long as we are consistent.
x=10
total = 0
if x!=0 :
    total +=x
print( f'the total is {total}')
```

```python
# lets learn about else if

number = int(input("Enter the number:"))
if number>0 :
    print( f' the number is positive')

else:
    print('The number is negative')
```

```python
# the elif statement works in the same way ,if any condition is true all the rest elif below statements are skipped 
number =20
if number>0 :
    print('Number is positive')
elif number<0 :
    print('Number is negative')
else :
    print('Number is neutral')
```

```python
#nested else if statement
number = -3 
if number>=0:
    if number==0:
        print('Number is 0')
    else :
        print('Number is positive')
else :
    print("number is negative")
```

```python
# lets learn about some more ways to write 
number = 10
if number>0 : print('Number is positive')

# we can also write in ternary method
grade = 70
result = 'pass' if grade>=50 else 'fail'

print(result)
```
