```python
# Input
account_balance= float(input("Enter your account balance:"))
risk_percent_per_trade=float(input("Enter Risk % per trade:"))
entry_price=float(input("Enter the entry price:"))
stop_loss=float(input("Enter the stop loss price"))

#Calculations
risk_amount = account_balance*risk_percent_per_trade*0.01
sl_distance = abs(stop_loss-entry_price)
position_size = risk_amount/sl_distance

#Output
print("Risk Amount:",risk_amount)
print("Stop Loss Distance:",sl_distance)
print("Number of units for entry:",position_size)
```
